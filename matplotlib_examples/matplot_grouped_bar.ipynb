### Grouped Bar Chart with Matplotlib
In this example, we compare scores between two groups (Men and Women) across five different categories (G1 to G5).

1. Import Dependencies:
First, we import matplotlib.pyplot for plotting and numpy to handle the numerical positioning of the bars.

```python
import matplotlib.pyplot as plt
import numpy as np
```

2. Define the Data: 
We define our categories and the corresponding values for each group.

```python
labels = ["G1", "G2", "G3", "G4", "G5"]
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # The label locations: [0, 1, 2, 3, 4]
width = 0.35  # The width of the bars
```

3. Create the PlotThis is the core logic. 

```python
fig, ax = plt.subplots()

# Create the bars
rects1 = ax.bar(x - width / 2, men_means, width, label="Men")
rects2 = ax.bar(x + width / 2, women_means, width, label="Women")

# Add text for labels, title and custom x-axis tick labels
ax.set_ylabel("Scores")
ax.set_title("Scores by group and gender")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()
```

4. Add Bar Labels and Fine-tune
We use bar_label to automatically annotate the height of each bar, making the chart much easier to read at a glance.

```python
# Add labels on top of the bars
ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

# Adjust layout to prevent clipping and show the plot
fig.tight_layout()

plt.show()
```
