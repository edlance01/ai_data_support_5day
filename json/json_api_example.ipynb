## JSON API Example

### Real-World JSON: Who is in Space right now?
In this section, we will:
* Make an HTTP GET request to a live API.
* Convert the JSON response into a Python list of dictionaries.
* Iterate through the data to display the names of astronauts currently in orbit.

### We’ll pull live data from an API. We'll use the Open Notify API, which provides real-time information about the International Space Station (ISS).

```python
import requests
import json

# The API endpoint for people currently in space
url = "http://api.open-notify.org/astros.json"

# 1. Get the data from the internet
response = requests.get(url)

# 2. Check if the request was successful (Status Code 200)
if response.status_code == 200:
    # .json() automatically parses the string into a Python Dictionary
    data = response.json()
    print("Successfully fetched live data!")
else:
    print(f"Failed to get data. Status code: {response.status_code}")
```

### Visualizing the Response
Before we extract names, let's see what the "raw" JSON looks like. APIs often nest data inside keys like people or results.

```python
# Let's 'pretty print' the response so we can see the structure
print(json.dumps(data, indent=2))

# Observation:
# The data has a 'number' key (total people)
# and a 'people' key which is a LIST of DICTIONARIES.
```

### Extracting Specific Information
Now that we know the structure, we can loop through the people list and print just the names and their respective spacecraft.

```python
print(f"There are currently {data['number']} humans in space:\n")

# Loop through the list of dictionaries
for person in data['people']:
    name = person['name']
    craft = person['craft']
    print(f"- {name} (onboard the {craft})")
```
