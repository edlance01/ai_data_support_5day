### Understanding JSON with Python

## JSON (JavaScript Object Notation) is the standard format for exchanging data. In Python, the json library handles the heavy lifting. There are four main methods to remember:

* dumps(): Dictionary to String.
* loads(): String to Dictionary.
* dump(): Dictionary to a File.
* load(): File to a Dictionary.

```python
import json

# Let's define some sample data
# Notice the Python-specific types: True (capitalized) and None
employee = {
    "id": 101,
    "name": "Sarah Connor",
    "department": "Engineering",
    "skills": ["Python", "Machine Learning", "Cybernetics"],
    "is_remote": True,
    "manager": None
}

print("Python Dictionary created.")
```

### Part 1: Serializing to a String
We use json.dumps() (the 's' stands for string) to turn our dictionary into a format that can be sent over the internet.

```python
# Convert Python dict to JSON string
# 'indent=4' makes it human-readable with nice spacing
employee_json = json.dumps(employee, indent=4)

print("Data converted to JSON string:")
print(employee_json)
print(f"\nNotice the type is now: {type(employee_json)}")
# Note: 'True' became 'true' and 'None' became 'null'
```

### Part 2: Parsing JSON back to Python
When you receive JSON data from an API, it arrives as a string. We use json.loads() (load string) to turn it back into a Python dictionary so we can access specific keys.

```python
# Let's take that string and turn it back into a Python object
data_from_api = json.loads(employee_json)

# Now we can access it like a normal dictionary
print(f"Employee Name: {data_from_api['name']}")
print(f"First Skill: {data_from_api['skills'][0]}")

if data_from_api["is_remote"]:
    print("This employee works from home.")
```

### Part 3: Working with Files
In most real-world scenarios, you'll be reading from a .json file. Here is how you write and read files directly.

```python
# Writing to a file
with open("files/data.json", "w") as f:
    json.dump(employee, f, indent=4)
print("File 'data.json' has been written.")

# Reading from a file
with open("files/data.json", "r") as f:
    file_content = json.load(f)

print("Successfully read from file:")
print(file_content["department"])
```
